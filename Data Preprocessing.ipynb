## Outlier Replacement - Median

```python
q1 = df.bmi.quantile(0.25)   #(Q1)
q3 = df.bmi.quantile(0.75)   #(Q3)
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
```

```python
Upper_limit = q3+1.5 * IQR
lower_limit = q1-1.5 * IQR
```

```python
Upper_limit
```

```python
df.median()
```

```python
df['bmi']=np.where(df['bmi']>Upper_limit,30,df['bmi'])
```
