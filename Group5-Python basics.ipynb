## statements

```python
## if


x=int(input('enter a number = '))

if x==100:
  print('This is my fav. number')
```

```python
## if else

x=int(input('enter a number = '))

if x==100:
  print('This is my fav. number')
else:
  print('This is not my fav. number')  
```

```python
## elif

x=int(input('enter a number = '))

if x<0:
  print('Negative')
elif x==0:
  print('Zero')  
else:
  print('Positive')  
```

## Loops

```python
list =['Appu','India','Ronak']

for i in list:
  print(i)
```

```python
# print first 15 elements

for i in range (15):
  print(i)
```

```python
# While loop

i = 0
while(i<15):
    print(i)
    i +=1     # i=i+1

```

```python
# print even numbers upto 20

i=2
while(i<=20):
  print(i)
  i+=2
```
